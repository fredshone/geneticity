```python
from geneticity import CityTemplate, City, Generation, CityEvolution
import matplotlib.pyplot as plt
```

```python
template = CityTemplate(name='Test', size = (12,12))
template.load_centric_city(centre_radius=2)
```

```python
city = City(template)
city.diagnose()
```

```python
gen = Generation(12,4,4,5)
gen.seed(city, shuffle=False)
```

```python
analysis = CityEvolution()
analysis.seed(gen)
```

```python
analysis.run()
```

```python
plt.plot(analysis.best_scores)
#plt.plot(analysis.population_scores)
#plt.plot(analysis.employer_scores)
plt.title("Most Fit Trend")
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.show()
```
