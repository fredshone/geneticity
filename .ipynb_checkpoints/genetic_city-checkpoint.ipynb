```python
def make_test_city(height, width, group):
    '''
    Make city dataframe from 
    '''
    headers = ['id', 'type', 'people', 'jobs', 'public']
    output = []
    ID = 1
    for y in range(height):
        for x in range(width):
            line = []
            # ID
            line.append(ID)
            # type - park, commercial, residential
            if y < group and x < group:
                line += ['residential', 100, 0, 0]
            elif y >= height - group and x >= width - group:
                line += ['commercial', 0, 100, 0]
            else:
                line += ['park', 0, 0, 100]
            output.append(line)
            ID += 1
    df = pd.DataFrame(output, columns=headers)
    df = df.set_index('id')
    return df
```

```python
from scipy.spatial.distance import cityblock
    
def calculate_block_utility(main_index, city):
    rtu = 0
    rdu = 0
    cpu = 0
    cau = 0
    block_id = city[main_index]
    stats = blocks.loc[block_id, :]
    # park - no util
    for index, block_id in np.ndenumerate(city): 
        block_stats = blocks.loc[block_id, :]
        dist = cityblock(main_index, index)     
        if dist == 0:  # Same block
            continue
        # resident work travel util - disutil for distance to work
        if dist <= 6:
            rtu += (1 - (dist/6)) * block_stats.jobs
            
        # resident density util - disutil for high density (max = 900)
        if dist < 2:
            rdu += 100 - block_stats.people
            
        # commercial - util for lots of close pop
        cpu += block_stats.people / dist
        
        # commercial - util for lots of close commercial
        cau += block_stats.jobs / (dist ** 2)
        
    r_score = ((rtu_weight * rtu) + (rdu_weight * rdu)) * stats.people
    c_score = ((cpu_weight * cpu) + (cau_weight * cau)) * stats.jobs
    score = (r_score * r_weight) + (c_score * c_weight)
    
    return np.array((score, r_score, c_score))
```

```python
def calc_city_utility(city):
    scores = np.array((0.,0.,0.))
    for index, _block in np.ndenumerate(city):
        scores += calculate_block_utility(index, city)
    return scores       
        
```

```python
import random

def recombine(mother, father):
    child = np.zeros_like(mother)
    for index, name in enumerate(blocks.index):
        if index < 50:
            position = np.where(mother == name)
        else:
            position = np.where(father == name)
            
        if child[position]:
            filtered = np.where(np.equal(child, 0))
            position = random.choice(list(zip(filtered[0], filtered[1])))

        child[position] = name

    return child
```

```python
import random

def mutate_chromosome(chromosome):
    first = (random.randrange(0, 10, 1), random.randrange(0, 10, 1))
    second = (random.randrange(0, 10, 1), random.randrange(0, 10, 1))

    chromosome[first[0], first[1]], chromosome[second[0], second[1]] = chromosome[second[0], second[1]], chromosome[first[0], first[1]]
    return chromosome
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def get_block(name):
    use = blocks.loc[name,'type']
    if use == 'park':
        return (use, 100)
    elif use == 'commercial':
        return (use, blocks.loc[name,'jobs'])
    elif use == 'residential':
        return (use, blocks.loc[name,'people'])
    else:
        print("Unknown type")
        
def get_comm(name):
    return blocks.loc[name,'jobs']

def draw_population(population, generation):
    get_ids = np.vectorize(get_comm)
    figure, axes = plt.subplots(1, 10, figsize=(20, 3))

    for index, city in enumerate(population):
        get_ids = np.vectorize(get_comm)
        ids = get_ids(city)
        axes[index].imshow(ids, interpolation="nearest", cmap=plt.cm.plasma)
        subplot_title = "Chromosome " + str(index)
        axes[index].set_title(subplot_title)

    plot_title = "Generation " + str(generation)
    figure.suptitle(plot_title)
    plt.show()
```

```python
import copy

most_fit = []
max_fitness = None
fittest_chromosome = None
```

```python
def get_cities_in_generation(generations, generation_num):
    # Get only the cities in this generation
    generation = list(filter(lambda k: k['generation'] == generation_num, 
                             generations))
    return sorted(generation, key=lambda k: k['fitness'][0])
```

```python
def sort_chromosomes(generation, most_fit):
    """
    Sort the chromosones in ascending order by fitness
    and update the most_fit list and the fittest_chromosome.
    """
    max_fitness = generation[-1]['fitness'][0]
    most_fit.append(max_fitness)
    fittest_chromosome = generation[-1]['chromosome']
    return most_fit, fittest_chromosome
```

```python
def copy_chromosomes(generation, generations):
    """
    Copy the surviving chromosomes in generation to a
    new generation and append this new generation to
    our generations list.
    """
    new_generation = copy.deepcopy(generation)

    new_generation[0]['active'] = False
    new_generation[1]['active'] = False
    new_generation[2]['active'] = False
    new_generation[3]['active'] = False

    active = list(filter(lambda k: k['active'] == True, new_generation))
    for chromosome in active:
        chromosome['generation'] += 1

    generations.extend(active)
    return generations
```

```python
def mate_chromosomes(generation, generations, generation_num):
    """
    Mate the two most fit chromosomes and add their
    children to the generation.
    """
    child = recombine(generation[-1]['chromosome'], generation[-2]['chromosome'])
    child = mutate_chromosome(child)
    generations.append({
      "active": True,
      "chromosome": child,
      "generation": generation_num + 1,
      "fitness": calc_city_utility(child)
    })
    child = mutate_chromosome(child)
    generations.append({
      "active": True,
      "chromosome": child,
      "generation": generation_num + 1,
      "fitness": calc_city_utility(child)
    })
    # Do this again for the second two most fit
    child = recombine(generation[-3]['chromosome'], generation[-4]['chromosome'])
    child = mutate_chromosome(child)
    generations.append({
      "active": True,
      "chromosome": child,
      "generation": generation_num + 1,
      "fitness": calc_city_utility(child)
    })
    child = mutate_chromosome(child)
    generations.append({
      "active": True,
      "chromosome": child,
      "generation": generation_num + 1,
      "fitness": calc_city_utility(child)
    })
    return generations
```

```python
def run_ga(input_generations):
    fittest = []
    max_fitness = None
    fittest_chromo = None
    for generation_num in range(300):
        generation = get_cities_in_generation(input_generations, generation_num)
        fittest, fittest_chromo = sort_chromosomes(generation, most_fit)

        input_generations = copy_chromosomes(generation, input_generations)
        input_generations = mate_chromosomes(generation, input_generations, generation_num)

        # Draw the generation that we just traversed through
        if generation_num % 15 == 0:
            draw_population(map(lambda d: d['chromosome'], get_cities_in_generation(input_generations, generation_num)), generation_num)

    return fittest, fittest_chromo
```

```python
most_fit, fittest_chromosome = run_ga(generations)
```

```python
plt.plot(most_fit)
plt.title("Most Fit Trend")
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.show()
```
