```python
from geneticity import CityTemplate, City, Generation, CityEvolution
import matplotlib.pyplot as plt
```

```python
template = CityTemplate(name='London', size = (10,10))
template.load_centric_city()
```

```python
city = City(template)
city.diagnose()
```

```python
template = CityTemplate(name='test', size = (10,10))
template.load_csv("city1.csv")
```

```python
city = City(template)
city.diagnose()
```

```python
gen = Generation(10,4,3,4)
gen.seed(city, shuffle='rows')
```

```python
analysis = CityEvolution()
analysis.seed(gen)
```

```python
analysis.run()
```
